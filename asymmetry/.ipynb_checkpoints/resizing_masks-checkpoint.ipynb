## Code used to resize masks and create a new folder with bigger black-background

```python
from PIL import Image
import os

def double_black_background(input_folder, output_folder):
    # Create the output folder 
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # list all files in the input folder
    files = os.listdir(input_folder)

    for file_name in files:
        if file_name.endswith('.png') or file_name.endswith('.jpg') or file_name.endswith('.jpeg'):
            # open the image using PIL
            image_path = os.path.join(input_folder, file_name)
            img = Image.open(image_path)

            # get the size of the original image
            width, height = img.size

            # calculate the dimensions for the new image with double black background
            new_width = width * 2
            new_height = height * 2

            # create a new image with double black background
            new_img = Image.new('RGB', (new_width, new_height), (0, 0, 0))

            # put the original image in the center of the new image
            offset_x = (new_width - width) // 2
            offset_y = (new_height - height) // 2
            new_img.paste(img, (offset_x, offset_y))

            # Save the new image to the output folder
            output_path = os.path.join(output_folder, file_name)
            new_img.save(output_path)

            print(f"Processed: {file_name}")

input_folder_path = 'masks'  
output_folder_path = 'masks'  

# Create a new folder for the processed images
new_output_folder_path = os.path.join(output_folder_path, 'processed_masks')
if not os.path.exists(new_output_folder_path):
    os.makedirs(new_output_folder_path)

double_black_background(input_folder_path, new_output_folder_path)
```
